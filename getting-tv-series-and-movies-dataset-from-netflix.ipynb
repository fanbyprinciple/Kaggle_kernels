```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200

df = pd.read_csv("../input/netflix-shows/netflix_titles.csv")
df.head()
```

# This notebook is to generate TV series and Movies dataset seperately from NETFLIX dataset

![](https://i.imgur.com/x3JQ0FD.png)

```python
df.describe()
```

```python
df["date_added"] = pd.to_datetime(df['date_added'])
df['year_added'] = df['date_added'].dt.year.astype('Int64')
df['month_added'] = df['date_added'].dt.month

df['duration'].head()
```

Lets try to identify between movies and tv series.

```python
#checking
[str(x).split(" ") for x in df['duration'][:5]]
```

```python
df['season_count'] = df.apply(lambda x : str(x['duration']).split(" ")[0] if "Season" in str(x['duration']) else "0", axis = 1)
df['duration'] = df.apply(lambda x : str(x['duration']).split(" ")[0] if "Season" not in str(x['duration']) else "0", axis = 1)
df.head()
```

```python
df["season_count"]
```

Now we have a way to work with only netflix movies only and movie tv series dataset.

```python
movies = df.loc[df["season_count"] == "0"]
movies.head()
```

```python
tv_series = df.loc[df["duration"] == "0"]
tv_series.head()
```

```python
movies.to_csv("movies_on_netflix.csv")
tv_series.to_csv("tv_series_on_netflix.csv")
```
