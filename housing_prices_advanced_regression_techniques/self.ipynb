```python
# loading the datasets
DATA_PATH = '../data/house_prices_advanced_regression_techniques'

import pandas as pd
import os

def load_housing_data(name):
    csv_path = os.path.join(DATA_PATH, name)
    return pd.read_csv(csv_path)
```

```python
housing_train = load_housing_data("train.csv")
housing_test = load_housing_data("test.csv")
housing_sample_output = load_housing_data("sample_submission.csv")
```

```python
dropset = []
for i in housing_train.columns:
    if (housing_train[i].dtype == 'O'):
        dropset.append(i)
        
```

```python
print(dropset)
```

```python
# dropping all the non numerical datset
housing_train_dropped = housing_train
for i in dropset:
    housing_train_dropped = housing_train_dropped.drop(i,axis=1)
    
```

```python
datatypes = housing_train_dropped.dtypes

print(datatypes.unique())
```

```python
# dropping rows with NaN values

#housing_train_dropped = housing_train_dropped.dropna()

# filling with 0
housing_train_dropped = housing_train_dropped.fillna(0)
```

```python
X = housing_train_dropped.drop("SalePrice", axis=1)
```

```python
# seperating labels and features in housing_train
y = housing_train_dropped["SalePrice"]
```

```python
y[:5]
```

```python
X[:5]
```

```python
from sklearn.svm import SVR

svm_reg = SVR(kernel="linear")
svm_reg.fit(X,y)
```

```python
# removing non integer columns

# dropping all the non numerical datset
housing_test_dropped = housing_test
for i in dropset:
    housing_test_dropped = housing_test_dropped.drop(i,axis=1)
    
```

```python
housing_test_dropped.head()
```

```python
housing_test_dropped['Id']
```

```python
# filling na with0
housing_test_dropped = housing_test_dropped.fillna(0)
```

```python
predictions = svm_reg.predict(housing_test_dropped)
```
